I am going to match all column names format. Gender and Tenure should start from the capital letter.

```python
df.rename(columns={'gender': "Gender", 'tenure' : "Tenure"}, inplace = True)
```

```python
df.head(2)
```

Lets create a column with tenure periods of 12 months. 1-12, 13-24,etc...

```python
print(df.Tenure.max())
print(df.Tenure.min())
```

```python
def tenure_p(month):
    if month in range(1,13):
        return '1-12'
    elif month in range(13,25):
        return '13-24'
    elif month in range(25,37):
        return '25-36'
    elif month in range(37,49):
        return '37-48'
    elif month in range(49,61):
        return '49-60'
    elif month in range(61,73):
        return '61-72'
```

```python
df['Tenure_periods'] = df['Tenure'].apply(tenure_p)
```
